```python
#importing libraries

from collections import Counter
from nltk.tokenize import sent_tokenize, word_tokenize
import pandas as pd
from pandas import DataFrame
```

```python
#Responsible person for Enterprise Architecture CSV, tokeinzation and printing it in list

RPs = []
with open ('revised_exported_EA_responsible_person.csv') as file_object:
    for trainline in file_object:
        tokens_train = sent_tokenize(trainline)
        RPs.extend(tokens_train)
    print(RPs)
```

```python
# Counter to count the number of times word repeated in EA CSV
counts = Counter(RPs)
print(counts)
```
